```python
# fig, ax = plt.subplots(1, 2, figsize=(10, 5), sharey=True)

plt.plot(steps_cnn, blur_cnn, label='Blur', color='red')
plt.plot(steps_cnn, blur_cnn, color='red', ls='--')
plt.plot(steps_d4, js_d4, label='JS', color='blue', ls='--')
plt.plot(steps_cnn, js_cnn, color='blue')
plt.plot(steps_cnn, max_distance_cnn, label='Max Distance', color = 'green')
plt.plot(steps_d4, max_distance_d4, color = 'green', ls='--')
plt.legend()
plt.yscale('log')
print('CNN mins', np.nanmin(blur_cnn), np.nanmin(js_cnn), np.nanmin(max_distance_cnn))
print('D4 mins', np.nanmin(blur_d4), np.nanmin(js_d4), np.nanmin(max_distance_d4))
print('CNN maxs', np.nanmax(blur_cnn), np.nanmax(js_cnn), np.nanmax(max_distance_cnn))
print('D4 maxs', np.nanmax(blur_d4), np.nanmax(js_d4), np.nanmax(max_distance_d4))
plt.title('MNIST-M, D4 (dashed) vs CNN (solid)')
plt.xlabel('Epochs')
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

# Plotting the losses for CNN
ax.plot(steps_cnn, train_class_loss_cnn, label='Classification Loss CNN', color='blue')
ax.plot(steps_cnn, train_domain_loss_cnn, label='Domain Loss CNN', color='red')

# Plotting the losses for D4
ax.plot(steps_d4, train_class_loss_d4, label='Classification Loss D4', color='blue', linestyle='--')
ax.plot(steps_d4, train_domain_loss_d4, label='Domain Loss D4', color='red', linestyle='--')

ax.legend()
ax.set_title('MNISTM Train Losses')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
plt.yscale('log')

plt.show()
```
